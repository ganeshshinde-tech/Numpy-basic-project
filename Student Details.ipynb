```python
import numpy as np
```

```python
#Project of Numpy
```

```python
marks = np.array(([40,50,40,50,40],
                [60,40,80,90,70],
                [100,70,80,90,60]))
```

```python
total_marks = np.sum(marks,axis = 1)
```

```python
Higest_marks = np.max(total_marks)
```

```python
Lowest_marks = np.min(total_marks)
```

```python
Percentage =(total_marks / 500 )* 100
```

```python
Class_Topper = np.max(Percentage)
```

```python
Topper_sub = np.max(marks, axis = 0)
```

```python
# Resulting the detials
```

```python
print("Marks of Students is :\n",marks),
print(" \nTotal Marks of Student is : ",total_marks),
print("Highest Marks of Students : ",Higest_marks),
print("Lowest Marks of Student : ",Lowest_marks),
print("Makrs of Topper is : ",Class_Topper),
print("The Percentages of Students is : ",Percentage),
print("Topper subject marks  : ",Topper_sub)
```
